```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# Define file paths
schools_complete_csv = Path("Resources/schools_complete.csv")
students_complete_csv = Path("Resources/students_complete.csv")


# Read School and Student Data File and store into Pandas DataFrames
schools_complete_df = pd.read_csv(schools_complete_csv)
students_complete_df = pd.read_csv(students_complete_csv)

# Combine the data into a single dataset.  
school_complete_df = pd.merge(schools_complete_df, students_complete_df, how="left", on=["school_name"])

school_complete_df
```

```python

# Calculate the total number of schools
total_schools = schools_complete_df['school_name'].nunique()

# Calculate the total number of students
total_students = students_complete_df['Student ID'].nunique()

# Calculate the total budget
total_budget = schools_complete_df['budget'].sum()

# Calculate the average math and reading scores
average_math_score = students_complete_df['maths_score'].mean()
average_reading_score = students_complete_df['reading_score'].mean()

# Calculate the percentage of students with passing math and reading scores
passing_math_percentage = (students_complete_df['maths_score'] >= 50).mean() * 100
passing_reading_percentage = (students_complete_df['reading_score'] >= 50).mean() * 100

# Calculate the percentage of students who passed both math and reading
overall_passing_percentage = ((students_complete_df['maths_score'] >= 50) & (students_complete_df['reading_score'] >= 50)).mean() * 100


# Dictionary 
summary_data = {
    'Total Schools': total_schools,
    'Total Students': total_students,
    'Total Budget': total_budget,
    'Average Math Score': average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_percentage]
}

# DataFrame 
summary_df = pd.DataFrame(summary_data)

# Summary DataFrame with formatted percentages
formatted_summary_df = summary_df.style.format({
    'Total Budget': '${:,.2f}',
    'Average Math Score': '{:.2f}',
    'Average Reading Score': '{:.2f}',
    '% Passing Math': '{:.2f}%',
    '% Passing Reading': '{:.2f}%',
    '% Overall Passing': '{:.2f}%'
})

formatted_summary_df



```
