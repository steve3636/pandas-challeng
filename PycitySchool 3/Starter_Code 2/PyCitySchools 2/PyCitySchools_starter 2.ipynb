## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# Define file paths
school_complete_csv = Path("Resources 2/school_complete.csv")
student_complete_csv = Path("Resources 2/student_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_complete_df = pd.read_csv(school_complete_csv)
student_complete_df = pd.read_csv(student_complete_csv)

# Merge school and student data
merged_df = pd.merge(school_complete_df, student_complete_df, how="left", on=["school_name"])
merged_df
```

```python
# Total number of schools
total_schools = merged_df['school_name'].nunique()

# Total number of students
total_students = merged_df['Student ID'].nunique()

# Total budget
total_budget = merged_df['budget'].sum()

# Average math and reading scores
average_math_score = merged_df['maths_score'].mean()
average_reading_score = merged_df['reading_score'].mean()

# % Students passing math score (50=>)
passing_math_percentage = (merged_df['maths_score'] >= 50).mean() * 100

# % of students passing reading score (50=>)
passing_reading_percentage = (merged_df['reading_score'] >= 50).mean() * 100

# % Students who passed both math and reading
overall_passing_percentage = ((merged_df['maths_score'] >= 50) & (merged_df['reading_score'] >= 50)).mean() * 100

# Summary dataframe in Box
summary_data = {
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_percentage]
}

summary_df = pd.DataFrame(summary_data)

# Format the displayed data
formatted_summary_df = summary_df.style.format({
    'Total Budget': '${:,.2f}',
    'Average Math Score': '{:.2f}',
    'Average Reading Score': '{:.2f}',
    '% Passing Math': '{:.2f}%',
    '% Passing Reading': '{:.2f}%',
    '% Overall Passing': '{:.2f}%'
})

# Display the formatted summary dataframe
formatted_summary_df
```

## School Summary

* Create a dataframe to hold the above results

```python
# Per Student Budget
merged_df['Per Student Budget'] = merged_df['budget'] / merged_df['size']

# assing indicators for math, reading, and overall passing
merged_df['Passing Math'] = merged_df['maths_score'] >= 50
merged_df['Passing Reading'] = merged_df['reading_score'] >= 50
merged_df['Overall Passing'] = merged_df['Passing Math'] & merged_df['Passing Reading']


# Rename columns
school_summary_df = school_summary_df.rename(columns={
    'school_name':'School Name',
    'type': 'School Type',
    'size': 'Total Students',
    'budget': 'Total School Budget',
    'maths_score': 'Average Math Score',
    'reading_score': 'Average Reading Score',
    'Passing Math': '% Passing Math',
    'Passing Reading': '% Passing Reading',
    'Overall Passing': '% Overall Passing'
})

# Custom formatting 
formatted_school_summary_df = school_summary_df.style.format({
    'Total School Budget': '${:,.2f}',
    'Per Student Budget': '${:,.2f}',
    'Average Math Score': '{:.2f}',
    'Average Reading Score': '{:.2f}',
    '% Passing Math': format_percentage,
    '% Passing Reading': format_percentage,
    '% Overall Passing': format_percentage
})

formatted_school_summary_df

```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Conditional statements
math_scores_9 = merged_df.loc[merged_df['year'] == 9].groupby('school_name')['maths_score'].mean()
math_scores_10 = merged_df.loc[merged_df['year'] == 10].groupby('school_name')['maths_score'].mean()
math_scores_11 = merged_df.loc[merged_df['year'] == 11].groupby('school_name')['maths_score'].mean()
math_scores_12 = merged_df.loc[merged_df['year'] == 12].groupby('school_name')['maths_score'].mean()

# Dataframe
average_math_scores_by_year = pd.DataFrame({
    
    'Year 9': math_scores_9,
    'Year 10': math_scores_10,
    'Year 11': math_scores_11,
    'Year 12': math_scores_12
})

# Displayed data
formatted_average_math_scores = average_math_scores_by_year.style.format({
    'Year 9': '{:.2f}',
    'Year 10': '{:.2f}',
    'Year 11': '{:.2f}',
    'Year 12': '{:.2f}'
})

formatted_average_math_scores
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 615, 645, 675]
spending_labels = ['<$585', '$585-615', '$615-645', '$645-675']

# DataFrame with the spending range
school_summary_df['Spending Ranges (Per Student)'] = pd.cut(
    school_summary_df['Per Student Budget'], spending_bins, labels=spending_labels, include_lowest=True
)

spending_summary = school_summary_df.groupby('Spending Ranges (Per Student)').agg({
    'Average Math Score': 'mean',
    'Average Reading Score': 'mean',
    '% Passing Math': 'mean',
    '% Passing Reading': 'mean',
    '% Overall Passing': 'mean'
}).reset_index()

def format_percentage(percentage):
    return f'{int(percentage * 100)}%'

formatted_spending_summary = spending_summary.style.format({
    'Average Math Score': '{:.2f}',
    'Average Reading Score': '{:.2f}',
    '% Passing Math': format_percentage,
    '% Passing Reading': format_percentage,
    '% Overall Passing': format_percentage
})

formatted_spending_summary

formatted_spending_summary
```

## Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']

school_summary_df['School Size'] = pd.cut(
    school_summary_df['Total Students'], size_bins, labels=size_labels, include_lowest=True
)

size_summary = school_summary_df.groupby('School Size').agg({
    'Average Math Score': 'mean',
    'Average Reading Score': 'mean',
    '% Passing Math': 'mean',
    '% Passing Reading': 'mean',
    '% Overall Passing': 'mean'
}).reset_index()

formatted_size_summary = size_summary.style.format({
    'Average Math Score': '{:.2f}',
    'Average Reading Score': '{:.2f}',
    '% Passing Math': format_percentage,
    '% Passing Reading': format_percentage,
    '% Overall Passing': format_percentage
})

formatted_size_summary
```
